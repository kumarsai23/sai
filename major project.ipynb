```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('data.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df
```

```python
df=df.drop(columns=['year','customer_id','phone_no'])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le=LabelEncoder()
```

```python
df['gender']=le.fit_transform(df['gender'])
```

```python
df['multi_screen']=le.fit_transform(df['multi_screen'])
```

```python
df['mail_subscribed']=le.fit_transform(df['mail_subscribed'])
```

```python
df
```

```python
test=df['churn']
train=df.drop(columns='churn')
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(train,test,test_size=0.25,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(random_state=42)
```

```python
rf.fit(xtrain,ytrain)
```

```python
preds=rf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(preds,ytest)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(preds,ytest)
```

```python
from sklearn.metrics import f1_score
f1_score(preds,ytest)
```

```python
from sklearn.metrics import recall_score
recall_score(preds,ytest)
```

```python
from sklearn.metrics import precision_score
precision_score(preds,ytest)
```
